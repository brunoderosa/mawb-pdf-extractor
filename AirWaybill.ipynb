```python
!pip install pandas PyPDF2

import os
import re
import pandas as pd
from PyPDF2 import PdfReader

pasta_pdfs = r""
arquivo_excel = r""

dados = []

for nome_arquivo in os.listdir(pasta_pdfs):
    if nome_arquivo.lower().endswith(".pdf"):
        caminho_arquivo = os.path.join(pasta_pdfs, nome_arquivo)
        reader = PdfReader(caminho_arquivo)
        pagina = reader.pages[0]
        texto = pagina.extract_text()

        match_mawb = re.search(r"\b\d{3}-\d{8}\b", texto)
        mawb = match_mawb.group() if match_mawb else "Não encontrado"

        match_dolar = re.search(
            r"EXCHANGE RATE USD[\s]?1\.00[\s]*X?[\s]*BRL[\s]*(\d+[\.,]\d+)", texto, re.IGNORECASE
        )
        valor_dolar = float(match_dolar.group(1).replace(',', '.')) if match_dolar else None

        dados.append({"Arquivo": nome_arquivo, "MAWB": mawb, "Valor dólar BRL": valor_dolar})

df = pd.DataFrame(dados)
df.to_excel(arquivo_excel, index=False)

print(f"Extração concluída. Dados salvos em {arquivo_excel}")
```
